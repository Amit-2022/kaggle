## EXPLORE DATA

```python
# SHAPE OF THE NEW DATA AFTER MERGE
df_merge.shape
```

```python
# Modify column ' genres ' , 'production_countries'
import ast
import random

def repair_array_bound_categories(arr):
    arr = ast.literal_eval(arr)
    
    if len(arr) == 0:
        return np.nan
    
    elif len(arr) == 1:
        return arr[0]
    
    else:
        return random.choice(arr)
        # return ":".join(arr)

df_merge["genres"]=df_merge["genres"].apply(repair_array_bound_categories)
df_merge["production_countries"]=df_merge["production_countries"].apply(repair_array_bound_categories)
```

```python
df_merge.head(5)
```

```python
# DELETE UNUSED COLUMNS
df_merge.drop(['description', 'age_certification', 'seasons', 'character'], axis=1, inplace=True)
```

```python
# Rows with empty values should be deleted
df_merge.dropna(inplace=True)
df_merge.info()
```

```python
# EXPLOREING DATA USING HISTOGRAM
df_merge.hist(figsize=(10, 10))
```

## Visulaization

```python
# What are the most popular types?
df_merge.groupby(['type']).sum().plot(kind='pie', y='tmdb_popularity', title='THE MOST POPULAR TYPES', colors=['red', 'yellow'], autopct='%1.0f%%')
```

```python
# the most popular genres?
df_merge.groupby(['genres']).sum().plot(kind='pie', y='tmdb_popularity', title='THE MOST POPULAR GENRES', figsize=(10, 10), autopct='%1.0f%%')
```

```python
# The Most Popular Type of Genres?
df_merge.groupby(['type', 'genres']).sum().plot(kind='bar', y='tmdb_popularity', title='THE MOST POPULAR TYPE OF GENRES', figsize=(10, 10), facecolor='b', edgecolor='k')
```
